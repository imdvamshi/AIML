```python
# load pandas inorder to use - pip install pandas
import pandas as pd
import numpy as np
```

## **Series**
### **A Series is a single column of data.**
### **It stores: values, index**

```python
a=np.array([10,20,30])
print(a)
```

```python
# creating a series
s = pd.Series(a)
print(s)
```

```python
print(type(s))
```

```python
# by default pandas gives indices from 0 in series
# you can also give custom own index names'
s = pd.Series(a,index=['first','second','third'])
print(s)
```

## accessing, modifying, slicing elements in series

```python
# accessing elements
print(s['first']) # print(s[0]) also works here
```

```python
# modifying elements
s['first'] = 20
print(s)
```

```python
# slicing elements
print(s[1:3])
```

```python
# +,-,*,/ on series
print(s)
print(s+5)
print(s-5)
print(s*5)
print(s/5)
```

```python
# creating series from dictionary
dict = {
    "virat" : 9000,
    "rohit" : 6000,
    "dhoni" : 5000
}
s = pd.Series(dict)
print(s)
```

## Important Series Attributes : index, values, dtype, shape

```python
print(s.index)
print(s.values)
print(s.dtype)
print(s.shape)
```

## Useful functions in Series : sum(), mean(), min(), max(), describe()

```python
print(s.sum())
print(s.mean())
print(s.min())
print(s.max())
```

```python
# describe() function - gives statistics
print(s.describe())
```

## **Filtering Data**

```python
data = {
    "name" : ['virat','rohit','dhoni'],
    "team" : ['rcb','mi','csk'],
    "runs" : [9000,6000,4000],
    "matches" : [255,200,200]
}
df = pd.DataFrame(data)
print(df)
```

```python
# conditional filtering
print(df['runs']>5000)
```

```python
print(df[df['runs']>5000])
```

```python
# multiple conditions [ &, |, ~]
print((df['team']=='rcb') & (df['runs']==9000))
```

```python
print(df[(df['team']=='rcb') & (df['runs']==9000)])
```

## **Handling Missing Data**

## Pandas represents missing values as:

## NaN ( Meaning: Not a Number )

```python
data = {
    "name": ["virat", "rohit", "dhoni", "hardik"],
    "runs": [9000, np.nan, 5000, 3000],
    "matches": [250, 220, np.nan, 180]
}
df = pd.DataFrame(data)
print(df)
# Why Decimal Appeared?
# Because: integers + NaN
# pandas converts column into float. Since NaN is internally float-based.
```

```python
# detecting null values - isnull()
print(df.isnull())
```

```python
# counting null values
print(df.isnull().sum())
```

```python
# opposite of isnull() is notnull()
print(df.notnull())
```

```python
# fillna() - used to replace missing values
print(df.fillna(0))
```

```python
# dropna() - remove rows with null values
print(df.dropna())
```

```python
### remove columns with null values - axis=1
print(df.dropna(axis=1))
```

```python
# fill null values with mean
mean = df['runs'].mean()
print(df.fillna(mean))
```

```python
# fill null values with median
median = df['runs'].median()
print(df.fillna(median))
```

```python
# forward fill - ffill -> replaces previous value with NaN
print(df.fillna(method="ffill"))
```

```python
# backward fill - bfill -> replaces next value with NaN
print(df.fillna(method='bfill'))
```

## **inplace=True -> permanently changes dataframe**

```python
# without inplace
print(df.fillna(0))
```

```python
print(df)
```

```python
# with inplace = True
df.fillna(1,inplace = True)
print(df)
```

## Problem: Extra Index Column

## By default pandas saves index also.

## Output file may look like:

## team,won

## 0,mi,5

## 1,csk,5

## Solution use index=False

```python
data.to_csv("updatediplwinsindexfalse.csv", index=False)
```
