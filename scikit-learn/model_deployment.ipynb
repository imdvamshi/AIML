## **Model Deployment**
### Model Deployment is the process of making a trained ML model available for real-world use.

```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame({
    'Area':[1000,1200,1500,1800,2000,2200,2500,2800,3000,3500],
    'Bedrooms':[2,2,3,3,3,4,4,4,5,5],
    'Price':[20,25,35,40,45,50,60,65,70,80]
})
data
```

```python
from sklearn.linear_model import LinearRegression
x = data[['Area','Bedrooms']]
y = data['Price']
model = LinearRegression()
model.fit(x.values,y)
prediction = model.predict([[2400,4]])
print(prediction)
```

## **TOPIC 1 : joblib**
### joblib is a library used to save and load trained machine learning models.

### Save a model using joblib - joblib.dump()

```python
# pip install joblib
import joblib
joblib.dump(model,'house_price_prediction.pkl') # .pkl means pickle file
```

### Load model using joblib - joblib.load()

```python
loaded_model = joblib.load('house_price_prediction.pkl')
```

```python
prediction = loaded_model.predict([[2400,4]])
print(prediction)
```

## **TOPIC 2 : pickle**

### Save a model using pickle

```python
import pickle
with open('house_model.pkl','wb') as file :
  pickle.dump(model,file)
```

### Load a model using pickle

```python
with open('house_model.pkl','rb') as file:
  loaded_model_ = pickle.load(file)
```

```python
prediction = loaded_model_.predict([[2400,5]])
print(prediction)
```

## joblib vs pickle
### Both save models. joblib is generally preferred for machine learning models.
### pickle - For General Python objects

### joblib - Optimized for ML models and large NumPy arrays

## **TOPIC 3 : DEPLOYMENT**

### FLASK

```python
from flask import Flask,request
import joblib
app = Flask(__name__)
model = joblib.load('house_price_prediction.pkl')
@app.route('/')
def home():
  return '''
  <h2>House Price Prediction</h2>
  <form action="/predict">

        Area:
        <input type="number" name="area"><br><br>

        Bedrooms:
        <input type="number" name="bedrooms"><br><br>

        <input type="submit" value="Predict">

    </form>
    '''
@app.route('/predict')
def predict():
  area = float(request.args.get('area'))
  bedrooms = int(request.args.get('bedrooms'))
  prediction = model.predict(
        [[area, bedrooms]]
    )
  return f"Predicted Price = {prediction[0]:.2f} Lakhs"
app.run(debug=True)
```

### FAST API
### It is an alternative to flask
### It is new,faster than flask and supports automatic api documentation

```python
from fastapi import FastAPI
import joblib
app = FastAPI()
model = joblib.load('house_price_prediction.pkl')
@app.get('/')
def home():
  return "house price prediction"
@app.get('/predict')
def predict():
  result = model.predict([[2400,4]])
  return f"Predicted Price: {result}"
```

### No app.run() in FastAPI
## Steps to run on website :
### 1) save file Save as : main.py
### 2) run from terminal : uvicorn main:app --reload
### where main=Python file name, app=FastAPI object, --reload=Automatically restart the server whenever the code changes.
