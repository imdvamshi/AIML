# **Hyper Parameter Tuning**
### A **hyperparameter** is a value that we set before training the model.

## Example:
### KNeighborsClassifier( n_neighbors=3 ) Here : n_neighbors = 3 is a Hyper parameter.

## More examples:
### 1) KNN -> n_neighbors, weights
### 2) Decision Tree -> max_depth, criterion
### 3) Random Forest - > n_estimators, max_depth
### 4) SVM -> C, kernel
### 5) K-Means -> n_clusters
### 6) PCA -> n_components

### **Tuning** The process of finding the best hyperparameter values for a model.

## **Grid Seach CV**

### Purpose : Find best hyperparameters automatically.

### Workflow : Parameter Grid -> Cross Validation -> Average Score -> Best Parameters

### Important : grid.best_params_, grid.best_score_

### Advantage : No manual trial-and-error.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
x = np.array([[1],[2],[3],[4],[8],[9],[10],[11]])
y = np.array([0,0,0,0,1,1,1,1])
model = KNeighborsClassifier()
params = {
   'n_neighbors' : [1,3,5]
}
grid = GridSearchCV(estimator=model,param_grid=params,cv=4)
grid.fit(x,y)
print(grid.best_params_) # best parameter value
print(grid.best_score_) # average cv accuracy across folds
best = grid.best_params_['n_neighbors']
model = KNeighborsClassifier(n_neighbors=best)
model.fit(x,y)
prediction = model.predict([[13]])
print(prediction)
```

```python
# multiple parameter example
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
x = np.array([[1],[2],[3],[4],[8],[9],[10],[11]])
y = np.array([0,0,0,0,1,1,1,1])
params = {
    'kernel' : ['linear','rbf','poly'],
    'C' : range(1,101)
}
grid = GridSearchCV(estimator=SVC(),param_grid=params,cv=4)
grid.fit(x,y)
print(grid.best_params_)
print(grid.best_score_)
model = SVC(C=grid.best_params_['C'],kernel=grid.best_params_['kernel'])
model.fit(x,y)
prediction = model.predict([[13]])
print(prediction)
```

## **Randomized Search CV**

### Purpose : Find best hyperparameters faster.

### Important Parameter : n_iter

### n_iter=10 -> Try 10 random combinations

## Difference:

### GridSearchCV : Finds the best parameter from all combinations.

### RandomizedSearchCV : Finds the best parameter only among the randomly sampled combinations.


### Advantage : Faster for large parameter spaces.

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
x = np.array([[1],[2],[3],[4],[8],[9],[10],[11]])
y = np.array([0,0,0,0,1,1,1,1])
model = RandomForestClassifier()
params = {
    "n_estimators" : list(range(1,101)),
}
random = RandomizedSearchCV(estimator=model,param_distributions=params,n_iter=10,cv=4)
random.fit(x,y)
print(random.best_params_)
print(random.best_score_)
model = RandomForestClassifier(n_estimators=random.best_params_['n_estimators'])
model.fit(x,y)
prediction = model.predict([[6]])
print(prediction)
```
