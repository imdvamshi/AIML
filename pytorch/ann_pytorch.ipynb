### **2. AutoGrad ( Automatic Differentiation )**
Autograd automatically performs the gradient calculations that Backpropagation requires. Calling loss.backward() computes the gradients for all trainable parameters(the tensors with this parameter : requires_grad=True will be tracked to compute gradients for it), and those gradients are stored in each parameter's .grad attribute

````python
x1 = torch.tensor(2.0, requires_grad=True)
y1 = x1**2
print(y1)
y1.backward()
print(x1.grad)
````

````python
x2 = torch.tensor(3.0, requires_grad=True)
y2 = 5*x2+2
print(y2)
y2.backward()
print(x2.grad)
````

### **7. FP until Loss function**

````python
import torch
import torch.nn as nn
class ann(nn.Module):
  def __init__(self):
    super().__init__()
    self.fc1 = nn.Linear(3,5)
    self.relu = nn.ReLU()
    self.fc2 = nn.Linear(5,3)
  def forward(self,x):
    x = self.fc1(x)
    x = self.relu(x)
    x = self.fc2(x)
    return x
model = ann()
x = torch.tensor([[2.5,1.8,3.2]])
actual = torch.tensor([1])
prediction = model(x)
print("prediction (logits) :")
print(prediction)
ce = nn.CrossEntropyLoss()
loss = ce(prediction,actual)
print("Loss:")
print(loss)
````

| Problem Type | Hidden Layer Activation | Output Layer Activation | Loss Function |
|--------------|-------------------------|-------------------------|---------------|
| Regression | `ReLU` | None | `nn.MSELoss()` |
| Binary Classification | `ReLU` | None *(Sigmoid applied internally)* | `nn.BCEWithLogitsLoss()` |
| Multi-Class Classification | `ReLU` | None *(Softmax applied internally)* | `nn.CrossEntropyLoss()` |

Note:
- ReLU (or another activation) is still used in the hidden layers.
- sigmoid or softmax were internally applied by bce or ce loss functions so there are not mentioned in output layer

### **8. Optimizers**

Note : torch.optim.SGD() can perform Batch Gradient Descent, Mini-Batch Gradient Descent, or Stochastic Gradient Descent depending on the batch size provided by the DataLoader.

````python
import torch
import torch.nn as nn
class ann(nn.Module):
  def __init__(self):
    super().__init__()
    self.fc1 = nn.Linear(3,5)
    self.relu = nn.ReLU()
    self.fc2 = nn.Linear(5,1)
  def forward(self,x):
    x = self.fc1(x)
    x = self.relu(x)
    x = self.fc2(x)
    return x
model = ann()
bce = nn.BCEWithLogitsLoss()
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)
x = torch.tensor([[2.5,1.8,3.2]])
actual = torch.tensor([[1.0]])
optimizer.zero_grad()
prediction = model(x)
loss = bce(prediction,actual)
print("loss :",loss)
loss.backward()
optimizer.step()
````

### **9. Training Loop**

````python
import torch
import torch.nn as nn
class ann(nn.Module):
  def __init__(self):
    super().__init__()
    self.fc1 = nn.Linear(3,5)
    self.relu = nn.ReLU()
    self.fc2 = nn.Linear(5,1)
  def forward(self,x):
    x = self.fc1(x)
    x = self.relu(x)
    x = self.fc2(x)
    return x
model = ann()
bce = nn.BCEWithLogitsLoss()
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)
x = torch.tensor([[2.5,1.8,3.2]])
actual = torch.tensor([[1.0]])
num_epochs = 20
for epoch in range(num_epochs):
  optimizer.zero_grad()
  prediction = model(x)
  loss = bce(prediction,actual)
  loss.backward()
  optimizer.step()
  print(f"Epoch {epoch+1} : Loss = {loss.item()}")
````

### **12. Train/Validation/Test**


| Dataset | Purpose | Updates Weights? | Used When? |
| :--- | :--- | :--- | :--- |
| **Training** | Learn | ✅ Yes | Every Epoch |
| **Validation** | Check Performance | ❌ No | Every Epoch (after training) |
| **Testing** | Final Evaluation | ❌ No | Once After Training |
| **Inference** | Predict New Data | ❌ No | After Deployment |


````python
import torch
import torch.nn as nn
from torch.utils.data import random_split,TensorDataset,DataLoader
class ann(nn.Module):
  def __init__(self):
    super().__init__()
    self.fc1 = nn.Linear(3,5)
    self.relu = nn.ReLU()
    self.fc2 = nn.Linear(5,1)
  def forward(self,x):
    x = self.fc1(x)
    x = self.relu(x)
    x = self.fc2(x)
    return x
model = ann()
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)
bce = nn.BCEWithLogitsLoss()
x = torch.tensor([
    [2.5,1.8,3.2],
    [1.2,4.5,2.7],
    [3.6,2.1,5.4],
    [6.3,7.8,1.5],
    [2.8,9.1,4.0],
    [5.4,2.3,1.8],
    [4.1,5.6,2.9],
    [7.2,3.5,6.4],
    [8.5,2.7,1.9],
    [3.3,6.6,5.2],
    [6.8,2.2,4.1],
    [5.7,7.4,3.8],
    [4.6,1.5,6.3],
    [8.4,5.1,2.2],
    [2.9,7.7,5.6]
])
y = torch.tensor([
    [1.0],
    [0.0],
    [1.0],
    [0.0],
    [1.0],
    [0.0],
    [1.0],
    [1.0],
    [0.0],
    [1.0],
    [1.0],
    [0.0],
    [1.0],
    [0.0],
    [1.0]
])
dataset = TensorDataset(x,y)
train, validation, test = random_split(dataset,[11,2,2]) # can use [0.8,0.1,0.1] percentage also
train_loader = DataLoader(train,batch_size=2,shuffle=True)
validation_loader = DataLoader(validation,batch_size=1,shuffle=False)
test_loader = DataLoader(test,batch_size=1,shuffle=False)
num_epochs = 20
for epoch in range(num_epochs):
  # training
  model.train()
  train_loss = 0
  for inputs,outputs in train_loader:
    optimizer.zero_grad()
    prediction = model(inputs)
    loss = bce(prediction,outputs)
    loss.backward()
    optimizer.step()
    train_loss += loss.item()
  train_loss /= len(train_loader)
  # validation
  model.eval()
  valid_loss = 0
  with torch.no_grad():
    for inputs,outputs in validation_loader:
      prediction = model(inputs)
      loss = bce(prediction,outputs)
      valid_loss += loss.item()
    valid_loss /= len(validation_loader)
  print(f"epoch {epoch+1} ")
  print("Training loss :",train_loss)
  print("Validation loss :",valid_loss)
  print("-"*40)
# testing
model.eval()
test_loss = 0
with torch.no_grad():
  for inputs,outputs in test_loader:
    prediction = model(inputs)
    loss = bce(prediction,outputs)
    test_loss += loss.item()
  test_loss /= len(test_loader)
print("Final Test Loss:",test_loss)
# inference -- new unseen data
new_data = torch.tensor([[4.9,3.1,6.5]])
model.eval()
with torch.no_grad():
  prediction = model(new_data)
  probability = torch.sigmoid(prediction)
print("Predicted class :",probability.item())
````


| Stage | `model.train()` | `model.eval()` | `torch.no_grad()` |
| :--- | :---: | :---: | :---: |
| **Training** | ✅ | ❌ | ❌ |
| **Validation** | ❌ | ✅ | ✅ |
| **Testing** | ❌ | ✅ | ✅ |
| **Inference** | ❌ | ✅ | ✅ |


### **13. Regularization Techniques**

Dropout

````python
import torch
import torch.nn as nn
from torch.utils.data import random_split,TensorDataset,DataLoader
class ann(nn.Module):
  def __init__(self):
    super().__init__()
    self.fc1 = nn.Linear(3,5)
    self.relu = nn.ReLU()
    self.dropout = nn.Dropout(p=0.5) ### dropout ----
    self.fc2 = nn.Linear(5,1)
  def forward(self,x):
    x = self.fc1(x)
    x = self.relu(x)
    x = self.dropout(x)
    x = self.fc2(x)
    return x
model = ann()
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)
bce = nn.BCEWithLogitsLoss()
x = torch.tensor([
    [2.5,1.8,3.2],
    [1.2,4.5,2.7],
    [3.6,2.1,5.4],
    [6.3,7.8,1.5],
    [2.8,9.1,4.0],
    [5.4,2.3,1.8],
    [4.1,5.6,2.9],
    [7.2,3.5,6.4],
    [8.5,2.7,1.9],
    [3.3,6.6,5.2],
    [6.8,2.2,4.1],
    [5.7,7.4,3.8],
    [4.6,1.5,6.3],
    [8.4,5.1,2.2],
    [2.9,7.7,5.6]
])
y = torch.tensor([
    [1.0],
    [0.0],
    [1.0],
    [0.0],
    [1.0],
    [0.0],
    [1.0],
    [1.0],
    [0.0],
    [1.0],
    [1.0],
    [0.0],
    [1.0],
    [0.0],
    [1.0]
])
dataset = TensorDataset(x,y)
train, validation, test = random_split(dataset,[11,2,2])
train_loader = DataLoader(train,batch_size=2,shuffle=True)
validation_loader = DataLoader(validation,batch_size=1,shuffle=False)
test_loader = DataLoader(test,batch_size=1,shuffle=False)
num_epochs = 20
for epoch in range(num_epochs):
  model.train()
  train_loss = 0
  for inputs,outputs in train_loader:
    optimizer.zero_grad()
    prediction = model(inputs)
    loss = bce(prediction,outputs)
    loss.backward()
    optimizer.step()
    train_loss += loss.item()
  train_loss /= len(train_loader)
  model.eval()
  valid_loss = 0
  with torch.no_grad():
    for inputs,outputs in validation_loader:
      prediction = model(inputs)
      loss = bce(prediction,outputs)
      valid_loss += loss.item()
    valid_loss /= len(validation_loader)
  print("epoch -- training loss -- validation loss")
  print(f"epoch {epoch+1} -- {train_loss} -- {valid_loss}")
  print("-"*40)
model.eval()
test_loss = 0
with torch.no_grad():
  for inputs,outputs in test_loader:
    prediction = model(inputs)
    loss = bce(prediction,outputs)
    test_loss += loss.item()
  test_loss /= len(test_loader)
print("Final Test Loss:",test_loss)
new_data = torch.tensor([[4.9,3.1,6.5]])
model.eval()
with torch.no_grad():
  prediction = model(new_data)
  probability = torch.sigmoid(prediction)
print("Predicted class :",probability.item())
````

````python
import torch
import torch.nn as nn
from torch.utils.data import random_split,TensorDataset,DataLoader
class ann(nn.Module):
  def __init__(self):
    super().__init__()
    self.fc1 = nn.Linear(3,5)
    self.relu = nn.ReLU()
    self.fc2 = nn.Linear(5,1)
  def forward(self,x):
    x = self.fc1(x)
    x = self.relu(x)
    x = self.fc2(x)
    return x
model = ann()
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)
bce = nn.BCEWithLogitsLoss()
x = torch.tensor([
    [2.5,1.8,3.2],
    [1.2,4.5,2.7],
    [3.6,2.1,5.4],
    [6.3,7.8,1.5],
    [2.8,9.1,4.0],
    [5.4,2.3,1.8],
    [4.1,5.6,2.9],
    [7.2,3.5,6.4],
    [8.5,2.7,1.9],
    [3.3,6.6,5.2],
    [6.8,2.2,4.1],
    [5.7,7.4,3.8],
    [4.6,1.5,6.3],
    [8.4,5.1,2.2],
    [2.9,7.7,5.6]
])
y = torch.tensor([
    [1.0],
    [0.0],
    [1.0],
    [0.0],
    [1.0],
    [0.0],
    [1.0],
    [1.0],
    [0.0],
    [1.0],
    [1.0],
    [0.0],
    [1.0],
    [0.0],
    [1.0]
])
dataset = TensorDataset(x,y)
train, validation, test = random_split(dataset,[11,2,2])
train_loader = DataLoader(train,batch_size=2,shuffle=True)
validation_loader = DataLoader(validation,batch_size=1,shuffle=False)
test_loader = DataLoader(test,batch_size=1,shuffle=False)
num_epochs = 20
best_val_loss = float("inf")
patience = 3
counter = 0
for epoch in range(num_epochs):
  model.train()
  train_loss = 0
  for inputs,outputs in train_loader:
    optimizer.zero_grad()
    prediction = model(inputs)
    loss = bce(prediction,outputs)
    loss.backward()
    optimizer.step()
    train_loss += loss.item()
  train_loss /= len(train_loader)
  model.eval()
  valid_loss = 0
  with torch.no_grad():
    for inputs,outputs in validation_loader:
      prediction = model(inputs)
      loss = bce(prediction,outputs)
      valid_loss += loss.item()
    valid_loss /= len(validation_loader)
  if valid_loss < best_val_loss : # early stopping ---
    best_val_loss = valid_loss
    counter = 0
  else :
    counter += 1
  if counter >= patience :
    print("Early stopping triggered")
    break
  print("epoch -- training loss -- validation loss")
  print(f"epoch {epoch+1} -- {train_loss} -- {valid_loss}")
  print("-"*40)
model.eval()
test_loss = 0
with torch.no_grad():
  for inputs,outputs in test_loader:
    prediction = model(inputs)
    loss = bce(prediction,outputs)
    test_loss += loss.item()
  test_loss /= len(test_loader)
print("Final Test Loss:",test_loss)
new_data = torch.tensor([[4.9,3.1,6.5]])
model.eval()
with torch.no_grad():
  prediction = model(new_data)
  probability = torch.sigmoid(prediction)
print("Predicted class :",probability.item())
````

````python
import torch
import torch.nn as nn
from torch.utils.data import random_split,Dataset,DataLoader

class DS(Dataset):
  def __init__(self):
    self.x = torch.tensor(data.drop(columns='Pass').values,dtype=torch.float32)
    self.y = torch.tensor(data['Pass'].values,dtype=torch.float32).unsqueeze(1)
  def __len__(self):
    return len(self.x)
  def __getitem__(self,index):
    return self.x[index],self.y[index]

class ann(nn.Module):
  def __init__(self):
    super().__init__()
    self.fc1 = nn.Linear(3,5)
    self.relu = nn.ReLU()
    self.fc2 = nn.Linear(5,1)
  def forward(self,x):
    x = self.fc1(x)
    x = self.relu(x)
    x = self.fc2(x)
    return x

model = ann()
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)
bce = nn.BCEWithLogitsLoss()

dataset = DS()
train,valid,test = random_split(dataset,[0.8,0.1,0.1])
train_loader = DataLoader(train,batch_size=3,shuffle=True)
valid_loader = DataLoader(valid,batch_size=1,shuffle=False)
test_loader = DataLoader(test,batch_size=1,shuffle=False)

num_epochs = 20
for epoch in range(num_epochs):

  model.train()
  train_loss = 0
  for inputs,outputs in train_loader:
    optimizer.zero_grad()
    prediction = model(inputs)
    loss = bce(prediction,outputs)
    loss.backward()
    optimizer.step()
    train_loss += loss.item()
  train_loss /= len(train_loader)

  model.eval()
  valid_loss = 0
  with torch.no_grad():
    for inputs,outputs in valid_loader:
      prediction = model(inputs)
      loss = bce(prediction,outputs)
      valid_loss += loss.item()
  valid_loss /= len(valid_loader)

  print(f"epoch {epoch+1} : Training_loss {train_loss}, Validation_loss {valid_loss}")
print("-"*30)

model.eval()
test_loss = 0
with torch.no_grad():
  for inputs,outputs in test_loader:
    prediction = model(inputs)
    loss = bce(prediction,outputs)
    test_loss += loss.item()
test_loss /= len(test_loader)
print(f"Test_loss {test_loss}")

model.eval()
cl = 0
x = torch.tensor([[2,60,40]],dtype=torch.float32)
with torch.no_grad():
  prediction = model(x)
  probability = torch.sigmoid(prediction)
if probability>=0.5:
  cl = 1
else :
  cl = 0
print(f"Predicted Class for {x} is {cl}")
````
