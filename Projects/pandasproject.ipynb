```python
import pandas as pd
import numpy as np
```

# **loading the data**

```python
data = pd.read_csv("students_performance.csv")
data.head()
```

# **understand the data**

```python
print(data.shape)
```

```python
print(data.dtypes)
```

# **Data Cleaning**

```python
print(data.isnull().sum())
```

```python
data=data.dropna()
```

```python
data
```

```python
print(data.isnull().sum())
```

```python
print(data.duplicated())
```

# **Explorative Data Analysis**

```python
# student who scored highest and least
data['total'] = data['Maths'] + data['Science'] + data['English']
data.head()
```

```python
# highest
data[data['total'] == data['total'].max()]
```

```python
# least
data[data['total'] == data['total'].min()]
```

```python
# average marks and subject wise avg
data['average'] = (data['Maths'] + data['Science'] + data['English'])/3
data.head()
```

```python
# Maths Average
print(data['Maths'].sum(axis=0))
```

```python
# Science Average
print(data['Science'].sum(axis=0))
```

```python
# English Average
print(data['English'].sum(axis=0))
```

```python
# relationship between studyhrs and marks
print(data['Study_Hours'].corr(data['total']))
```

```python
# relationship betweem studyhrs and pass/fail
data['Final_Result']=data['Final_Result'].map({'Pass':1,'Fail':0})
data.head()
```

```python
print(data['Study_Hours'].corr(data['Final_Result']))
```

```python
# Sports students performing better?
data.groupby('Sports')['total'].mean()
```

```python
# attendance affecting results
print(data['Attendance'].corr(data['Final_Result']))
```

# **Statistical Analysis**

```python
math_marks = np.array(data['Maths'])
science_marks = np.array(data['Science'])
english_marks = np.array(data['English'])
study_hrs = np.array(data['Study_Hours'])
```

```python
print("Average of math marks:",math_marks.mean())
print("Average of science marks:",science_marks.mean())
print("Average of english marks:",english_marks.mean())
print("Average Study Hours:",study_hrs.mean())
```
