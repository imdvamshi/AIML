```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('srh vs rcb final 2016.csv')
dataset
```

```python
sns.set_style('ticks')
```

# **lineplot**

```python
plt.figure(figsize=(8,5),dpi=100)
plt.title('Worm',fontdict={'fontname':'monospace','fontsize':20})
sns.lineplot(data=dataset,x='Over',y='Total_Score',hue='Innings',palette={'RCB':'red','SRH':'orange'})
srhd = dataset[dataset['Innings']=='SRH']
srhw = srhd[srhd['Wickets'].diff()>0]
rcbd = dataset[dataset['Innings']=='RCB']
rcbw = rcbd[rcbd['Wickets'].diff()>0]
sns.scatterplot(data=srhw,x='Over',y='Total_Score',s=50,color='orange')
sns.scatterplot(data=rcbw,x='Over',y='Total_Score',s=50,color='red')
plt.xticks(list(range(1,21)))
plt.legend()
plt.savefig('worm 2016 ipl final.png',dpi=300)
plt.show()
```

# **barplot**

```python
sns.barplot(data=dataset,x='Over',y='Runs_This_Over',hue='Innings',palette={'SRH':'orange','RCB':'RED'})
plt.show()
```

# **Histogram**

```python
sns.histplot(data=dataset,x='Runs_This_Over')
plt.show()
```

```python
sns.histplot(data=dataset,x='Runs_This_Over',kde=True)
plt.show()
```

# **count plot**

```python
sns.countplot(data=dataset,x='Runs_This_Over',color='#abcdef')
plt.show()
```

```python
sns.scatterplot(data=dataset,x='Over',y='Runs_This_Over',hue='Innings',palette={'SRH':'Orange','RCB':'red'})
plt.show()
```
