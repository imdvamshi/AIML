```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('srh vs rcb final 2016.csv')
data
```

```python
srh = data[data['Innings'] == 'SRH']
rcb = data[data['Innings'] == 'RCB']
```

```python
srh.head()
```

```python
rcb.head()
```

# **Line Plot**

```python
plt.figure(figsize=(8,5),dpi=100)
plt.plot(srh['Over'],srh['Runs_This_Over'],'.-',color='orange',label='SRH')
plt.plot(rcb['Over'],rcb['Runs_This_Over'],'r.-',label='RCB')
plt.title('overs comparison chart',fontdict={'fontname':'monospace','fontsize':20})
plt.xlabel('Over')
plt.ylabel('Runs')
plt.xticks(list(range(1,21)))
plt.legend()
plt.savefig('overs comparison chart.png',dpi=300)
plt.show()
```

# **Bar Chart**

```python
plt.figure(figsize=(8,5),dpi=100)
plt.bar(srh['Over']-0.2,srh['Runs_This_Over'],width=0.4,color='orange',label='SRH')
plt.bar(rcb['Over']+0.2,rcb['Runs_This_Over'],width=0.4,color='red',label='RCB')
plt.title('Runs Scored Per Over',fontdict={'fontname':'monospace','fontsize':20})
plt.xlabel('Over')
plt.ylabel('Runs')
plt.xticks(list(range(1,21)))
plt.yticks(list(range(2,22,2)))
plt.legend()
plt.savefig('runs scored per over.png',dpi=300)
plt.show()
```

# **Histogram**

```python
runs = data['Runs_This_Over']
plt.figure(figsize=(8,5),dpi=100)
plt.hist(runs,color='#abcdef')
plt.title('Distribution of runs per over',fontdict={'fontname':'monospace','fontsize':20})
plt.xlabel('runs scored in a over')
plt.ylabel('no of times scored')
plt.savefig('distribution of runs per over.png',dpi=300)
plt.show()
```

# **Scatter Plot**

```python
plt.figure(figsize=(8,5),dpi=100)
plt.scatter(srh['Over'],srh['Runs_This_Over'],color='orange',label='SRH')
plt.scatter(rcb['Over'],rcb['Runs_This_Over'],color='Red',label='RCB')
plt.title('over vs runs',fontdict={'fontname':'monospace','fontsize':20})
plt.xlabel('Over')
plt.ylabel('Runs')
plt.xticks(list(range(1,21)))
plt.savefig('over vs runs.png',dpi=300)
plt.show()
```

# **Pie Chart**

```python
plt.figure(figsize=(8,5),dpi=100)
labels = ['SRH', 'RCB']
values = [srh['Runs_This_Over'].sum(),rcb['Runs_This_Over'].sum()]
plt.pie(values, labels=labels, colors=['orange','red'])
plt.title('Share of Total Match Runs')
plt.savefig('share of total match runs.png',dpi=300)
plt.show()
```

# **Sub Plot**

```python
plt.figure(figsize=(8,5),dpi=100)
plt.subplot(1,2,1)
plt.plot(srh['Over'],srh['Runs_This_Over'],'.-',color='orange',label='SRH')
plt.xlabel('Over')
plt.ylabel('Runs')
plt.legend()
plt.subplot(1,2,2)
plt.plot(rcb['Over'],rcb['Runs_This_Over'],'r.-',label='RCB')
plt.xlabel('Over')
plt.ylabel('Runs')
plt.legend()
plt.savefig('overs comparison sidebyside.png',dpi=300)
plt.show()
```
