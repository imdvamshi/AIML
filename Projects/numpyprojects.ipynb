## **Students Marks Analyzer**
### Create a NumPy array of marks for 5 students in 3 subjects.

### **Find:**

### Total marks of each student

### Average marks

### Highest scorer

### Subject-wise average

```python
import numpy as np
marks = np.array([
    [78, 85, 90],
    [67, 72, 70],
    [90, 92, 88],
    [56, 60, 65],
    [80, 75, 85]
])
print(marks)
```

```python
# total marks of each student
b = np.sum(marks,axis=1)
print("Total marks of student 1 :",b[0])
print("Total marks of student 2 :",b[1])
print("Total marks of student 3 :",b[2])
print("Total marks of student 4 :",b[3])
print("Total marks of student 5 :",b[4])
```

```python
# average marks
c = np.mean(marks,axis=1)
print("Average marks of student 1 :",c[0])
print("Average marks of student 2 :",c[1])
print("Average marks of student 3 :",c[2])
print("Average marks of student 4 :",c[3])
print("Average marks of student 5 :",c[4])
```

```python
# highest scorer
d = np.argmax(b)
print("Highest scorer is student", d + 1)
```

```python
# subject wise average
e = np.mean(marks,axis=0)
print("Average marks of subject 1:",e[0])
print("Average marks of subject 2:",e[1])
print("Average marks of subject 3:",e[2])
```

## **Weather Temperature Analysis**
### Store temperatures of 7 days.

### **Find**:

### Hottest day

### Coldest day

### Average temperature

### Days above average

```python
import numpy as np
temp = np.array([32, 35, 30, 31, 36, 38, 33])
print(temp)
```

```python
# hotest day
print("Hotest day :",np.argmax(temp)+1,"with a temp of",np.max(temp))
```

```python
# coldest day
print("Coldest day :",np.argmin(temp)+1,"with a temp of",np.min(temp))
```

```python
# average temperature
print("Average temperature:",np.mean(temp))
```

```python
# days above average
print(temp[temp>np.mean(temp)])
```

## **AI/ML Style Dataset Normalization**
### Normalize values between 0 and 1.


```python
import numpy as np
data = np.array([10, 20, 30, 40, 50])
normalised = (data-np.min(data))/(np.max(data)-np.min(data))
print(normalised)
```

## **Attendance Checker**
### 1 means Present
### 0 means Absent

### Total present days

## **Dice Simulation**
### Simulate rolling a dice 20 times.

### Count how many times each number appears.

```python
import numpy as np
dice=np.random.randint(1,7,20)
print("dice roll for 20 times:",dice)
count=np.bincount(dice)
print("count of how many times each no appeared:",count[1:])
```
