## **SQL** - Structured Query Language is the language( commands )

## **SQLite, MySQL, PostgreSQL, and Oracle** are database systems that store data and execute SQL commands.



> SQLite3 = Python's simplest database system that stores data in a .db file and executes SQL commands.


> sqlite3 is Python's module (library) for talking to SQLite database.





# **BASICS**

```python
import sqlite3 # The sqlite3 module comes pre-installed with Python, so no pip install is needed.
import pandas as pd
```

# **Step 1 : Connect**

```python
conn = sqlite3.connect('company.db')
```

# **Step 2 : Create a table**

```python
conn.execute("CREATE TABLE Employees(id INTEGER, name TEXT, salary INTEGER)")
```

# **Step 3 : Insert Values**

```python
conn.execute("""
    INSERT INTO Employees VALUES(1,'vamshi',30000),(2,'Abhi',28000),(3,'Varshith',50000)
""")
```

# **Step 4 : Save Changes**
### Use commit after:

### CREATE, INSERT, UPDATE, DELETE

```python
conn.commit()
```

# **Step 5 : Read data back**
## To Display rows using Pandas DataFrame

```python
df = pd.read_sql_query("SELECT * FROM EMPLOYEES",conn)
```

```python
df
```

## **Display Rows Without DataFrame**

```python
conn.execute('SELECT * FROM Employees') # runs the query, but it does not display the results.
```

```python
cursor = conn.execute("SELECT * FROM Employees")
for row in cursor:
  print(row)
```

## **fetchone()** - Gets one row at a time.
## **fetchall()** - Gets all remaining rows.
## **fetchmany()** - Gets a specific number of rows.

```python
cursor = conn.execute("SELECT * FROM Employees;")
```

```python
# fetchone()
row = cursor.fetchone()
print(row)
```

```python
# Calling fetchone() Again - Cursor remembers its position.
row = cursor.fetchone()
print(row)
```

```python
cursor = conn.execute("SELECT * FROM Employees")
```

```python
# fetchmany()
rows = cursor.fetchmany(2)
print(rows) # returns list of rows
```

```python
# fetchall()
r = cursor.fetchall()
print(r)
```

# **Parameterized Queries (? Placeholders)**

```python
conn.execute("CREATE TABLE tutorial(name TEXT,age INTEGER)")
```

```python
# When inserting values from Python, don't build SQL strings manually. - works but not a good practice
name = 'vamshi'
age = 22
conn.execute(f"INSERT INTO tutorial VALUES('{name}',{age})")
```

```python
# Use ? Placeholders
name = 'dunna'
age = 16
conn.execute("INSERT INTO tutorial VALUES(?,?)",(name,age))
```

```python
df = pd.read_sql_query("SELECT * FROM tutorial" ,conn)
df
```

```python
# using placeholders in where clause
age = 20
df = pd.read_sql_query("SELECT * FROM tutorial WHERE age > ?",conn,params=(age,)) #The comma is important.
# A single-value tuple in Python needs a trailing comma, so (age,) written.
df
```

# **Write DataFrame to SQLite**

```python
df = pd.DataFrame({'ID':[1,2,3],'SALARY':[10000,45000,25000],'DEPT':['SC','HR','IT']})
df
```

### Meaning of parameters used in df.to_sql()
### if_exists="replace" - Delete old table and create new one.
### if_exists="append" - Add rows to existing table.
### index=False - Don't save Pandas index as a column.

```python
#### df.to_sql()
df.to_sql("salary",conn,if_exists="replace",index=False)
```

```python
c = conn.execute("SELECT * FROM salary")
for row in c:
  print(row)
```
