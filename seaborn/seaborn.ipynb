```python
# install and load before use - pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# **seaborn provides built-in datasets**

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

# **Themes and Styles**
## Theme -> Its like setting theme in your phone [ Syntax : sns.set_theme() ]

## Style -> Its like setting wallpaper in your phone [ Syntax : sns.set_style() ]

```python
# Theme
sns.set_theme(
    context='talk', # big text
    style='white', # white background
    font='serif', # fancy font
    palette='bright') # changes color of bars/dots
sns.scatterplot(data=tips,x='total_bill',y='tip')
plt.show()
```

```python
# Style - (whitegrid, darkgrid, white, dark, ticks)
sns.set_style('whitegrid')
sns.scatterplot(data=tips,x='total_bill',y='tip')
plt.show()
```

```python
sns.set_style('darkgrid')
sns.scatterplot(data=tips,x='total_bill',y='tip')
plt.show()
```

```python
sns.set_style('white')
sns.scatterplot(data=tips,x='total_bill',y='tip')
plt.show()
```

```python
sns.set_style('dark')
sns.scatterplot(data=tips,x='total_bill',y='tip')
plt.show()
```

```python
sns.set_style('ticks')
sns.scatterplot(data=tips,x='total_bill',y='tip')
plt.show()
```

# **3) Bar Plot**
## A Seaborn barplot is a visual tool that displays a summary metric (usually the average) for different categories
## special features :

## estimator -> automatic math

## error bar -> black line on bars (Short Line = Predictable,Tall Line = Unpredictable )



```python
sns.barplot(data=tips,x='day',y='total_bill')
plt.show()
```

```python
# without errorbar
sns.barplot(data=tips,x='day',y='total_bill',errorbar=None)
plt.show()
```

```python
# using estimator - automatic math
sns.barplot(data=tips,x='day',y='total_bill',estimator=np.sum) # np.median,np.max etc.
plt.show()
```

```python
# using hue
sns.barplot(data=tips,x='day',y='total_bill',hue='sex')
plt.show()
```

# **6) Box Plot**
## Find outliers and spread.
## Extremely Important in ML. Used for:

## Outlier detection
## Data cleaning

boxplot-quartiles.webp



> Every box on this chart is broken into 4 parts:The Line Inside the Box: This is the median (true middle). Exactly half of the customers spent less than this line, and half spent more.


> The Box Itself: This holds the middle 50% of all customers. It shows the typical, everyday spending range.


> The T-Shaped Lines (Whiskers): These show the highest and lowest normal bills on that day.


> The Floating Circles at the Top: These are outliers. They represent rare, unusually expensive bills (like a massive $50 party bill) that do not match regular customer habits.





```python
sns.boxplot(data=tips,x='day',y='total_bill')
plt.show()
```

# **8) Heat Map**
## Visualize correlations.

## Create correlation matrix

```python
corr = tips.corr(numeric_only=True)
sns.heatmap(corr)
plt.show()
```

```python
# displaying values on cells
corr = tips.corr(numeric_only=True)
sns.heatmap(corr,annot=True)
plt.show()
```

# **9) Pair Plot**
## Shows relationship between ALL numerical columns.
## Automatically creates: Histograms, Scatterplots


> For \(n\) variables, a pairplot creates an \(n \times n\) grid containing \(n^{2}\) total figures.It shows \(n\) diagonal charts for individual distributions ( While a histogram is the most common choice ) and \(n(n - 1)\) off-diagonal scatter plots for relationships.


> For 3 variables, a pairplot creates a 3x3 grid containing 9 total figures.It shows 3 diagonal charts for individual distributions( While a histogram is the most common choice ) and 6 off-diagonal scatter plots for relationships.


```python
sns.pairplot(tips)
plt.show()
```
