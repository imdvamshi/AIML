## Property	Meaning                       

## a.size->Number of elements

## a.itemsize->Memory per element

## a.nbytes->Total memory used

```python
x=np.array([1,2,3],dtype='int16')
y=np.array([1,2,3],dtype='int32')
z=np.array([1,2,3],dtype='int64')#can also be written as z=np.array([1,2,3],dtype=np.int64)
```

```python
print(x.size)
print(y.size)
print(z.size)
```

```python
#1byte=8bits so2bytes=16bit,4bytes=32bit,8bytes=64bit
print(x.itemsize)
print(y.itemsize)
print(z.itemsize)
```

```python
print(x.nbytes)
print(y.nbytes)
print(z.nbytes)
```

## Be careful when copying arrays!!!

```python
a=np.array([1,2,3])
print(a)
```

```python
b=a
print(b)
```

```python
#b becomes another reference (another name) to the SAME array.
#So both variables point to the same memory.
#(this topic is applicable to all mutable objs like list,dict,sets,numpy array)
b[0]=10
print(b)
print(a)
```

```python
# use copy() to create an independent object
c=a.copy()
print(a)
c[1]=20
print(c)
print(a)
```

## **Linear Algebra**

```python
# to find dot product
a=np.array([1,2,3])
b=np.array([4,5,6])
print(np.dot(a,b))
```

```python
# if a=mXn matrix and b=nXp matrix -> then matrix multiplication is possible
# resultant matrix=mXp
a=np.ones((2,3))
b=np.full((3,2),3)
# a*b gives error here
print(np.matmul(a,b))
```

```python
# to find transpose
a=np.array([[1,2],[3,4]])
print(a)
print(np.transpose(a))
```

```python
# to find determinant
a=np.array([[1,2],[3,4]])
print(np.linalg.det(a))
```

```python
b=np.identity(3) # det(Identity matrix)=always 1
print(b)
print(np.linalg.det(b))
```

```python
# to find inverse of a matrix
a=np.array([[1,2],[3,4]])
print(np.linalg.inv(a))
```

```python
# to find eigen values and eigen vectors
a=np.array([[2,0],[0,3]])
print(np.linalg.eig(a))
```

```python
# singular vector decomposition
a=np.array([[1,2],[3,4]])
print(np.linalg.svd(a))
```

```python
# using statistics module
import statistics
a=np.array([1,2,3])
print(statistics.mean(a))
print(statistics.median(a))
print(statistics.mode(a))
print(statistics.variance(a))
print(statistics.stdev(a.tolist()))
```

```python
# using numpy module
import numpy as np
b=np.array([1,2,3])
print(np.mean(b))
print(np.median(b))
# no function for mode in numpy
print(np.var(b))
print(np.std(b))
```

## **Reorganizing Arrays**

```python
# reshape()
a=np.array([[1,2,3],[4,5,6]])
print(a.shape)
print(a)
a=a.reshape((3,2))
print(a.shape)
print(a)
```

```python
# vertically stacking vectors
a=np.array([1,2,3])
b=np.array([4,5,6])
c=np.vstack((a,b))
# c=np.vstack([a,b]) also works
print(c)
```

```python
# horizontally stacking vectors
a=np.array([[1,1,1],[1,1,1]])
b=np.array([[0,0,0,0],[0,0,0,0]])
c=np.hstack([a,b])
print(c)
```

```python
a=np.array([1,2,3,4])
print(a[[1,3]])
```
