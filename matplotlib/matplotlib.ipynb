```python
# load matplotlib first inorder to use - pip install matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# **Line plot**
## A line plot is a graph that connects data points with lines to show trends or changes over time.
## Syntax : plt.plot(x, y)
## Use : It is used to visualize trends, patterns, or relationships between two variables.

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y)

plt.show()
```

```python
# add a title to graph
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.plot(x, y)

plt.title("Our First Graph", fontdict = {'fontname':'monospace', 'fontsize':20})

plt.show()
```

```python
# add x and y labels
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.plot(x, y)
plt.title("Our First Graph", fontdict = {'fontname':'monospace', 'fontsize':20})

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```

```python
# add grid
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.plot(x, y)
plt.title("Our First Graph", fontdict = {'fontname':'monospace', 'fontsize':20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.grid() #optional

plt.show()
```

```python
# ticks marks of graph (scale of graph)
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.plot(x, y)
plt.title("Our First Graph", fontdict = {'fontname':'monospace', 'fontsize':20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

plt.show()
```

```python
# legend (1st way)
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.plot(x, y, label = '2x')
plt.title("Our First Graph", fontdict = {'fontname':'monospace', 'fontsize':20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

plt.legend()

plt.show()
```

```python
# legend (2nd way)
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.plot(x, y)
plt.title("Our First Graph", fontdict = {'fontname':'monospace', 'fontsize':20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

plt.legend(['2x'], loc="lower right")
# plt.legend(['2x']) default also works

plt.show()
```

```python
# customize line and marker
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='-', markersize=10, markeredgecolor='red')

plt.title("Our First Graph", fontdict = {'fontname':'monospace', 'fontsize':20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])
plt.legend()
plt.show()
```

```python
# shorthand customize
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# format = '[color][marker][line]'
plt.plot(x, y, 'b*--', label = '2x')

plt.title("Our First Graph", fontdict = {'fontname':'monospace', 'fontsize':20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])
plt.legend()
plt.show()
```

```python
# resize graph
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.figure(figsize = (8,5), dpi = 100)
# (8,5) means 8 inches wide and 5 inches tall
# dpi specifies pixels per inch

plt.plot(x, y, 'r.-', label = '2x')
plt.title("Our First Graph", fontdict = {'fontname':'monospace', 'fontsize':20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])
plt.legend()
plt.show()
```

```python
# multiple lines
x = [1,2,3,4,5]
y1 = [2,4,6,8,10]
y2 = [1,3,6,9,12]
plt.figure(figsize = (8,5), dpi = 100)
plt.plot(x,y1, 'r.-', label = 'Team A')
plt.plot(x,y2, 'y.-', label = 'Team B')
plt.title('TeamA vs TeamB', fontdict = {'fontname':'monospace', 'fontsize':20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.show()
```

```python
# save figure
x = [1,2,3,4,5]
y1 = [2,4,6,8,10]
y2 = [1,3,6,9,12]
plt.figure(figsize = (8,5), dpi = 100)
plt.plot(x,y1, 'r.-', label = 'Team A')
plt.plot(x,y2, 'y.-', label = 'Team B')
plt.title('TeamA vs TeamB', fontdict = {'fontname':'monospace', 'fontsize':20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([1,2,3,4,5])
plt.yticks([2,4,6,8,10,12])
plt.legend()

# dpi 300 is good when saving so graph has high resolution
plt.savefig("my_firstgraph.png",dpi=300)

plt.show()
```

# **Bar Chart**
## A bar chart is a graph that represents data using rectangular bars, where the height (or length) of each bar shows the value.
## Syntax : plt.bar(x, y)
## Use: It is used to compare values across different categories.

```python
labels = ['A','B','C']
values = [2,4,3]

plt.bar(labels, values)

plt.show()
```

```python
# designing bars using set_hatch()
labels = ['A','B','C']
values = [2,4,3]
bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')

plt.show()
```

```python
# resizing and then saving as img
labels = ['A','B','C']
values = [2,4,3]
plt.figure(figsize = (8,5), dpi = 100)
bars = plt.bar(labels, values)
plt.title('Our First Bar Chart', fontdict = {'fontname':'monospace', 'fontsize':20})
bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')
plt.savefig('my_firstbarchart.png',dpi = 300)
plt.show()
```

# **Histogram**
## A histogram is a graph that groups numerical data into ranges (bins) and shows how frequently values occur in each range.
## Syntax : plt.hist(data)
## Use: It is used to visualize the distribution and frequency of numerical data.

```python
marks = [50,60,70,80,90,90,95,100,40,50]

plt.hist(marks, color='#abcdef')

plt.show()
```

# **Scatter Plot**
## A scatter plot is a graph that displays individual data points to show the relationship between two numerical variables.
## Syntax : plt.scatter(x, y)
## Use: It is used to identify relationships, correlations, patterns, or outliers between two variables.

```python
x = [1,2,3,4,5]
y = [25,45,55,65,75]

plt.scatter(x,y)

plt.show()
```

# **Pie Chart**
## A pie chart is a circular graph divided into slices, where each slice represents a category's proportion of the whole.
## Syntax : plt.pie(values, labels=labels)
## Use: It is used to show the percentage or proportion of different categories in a dataset.

```python
sizes = [40,30,20,10]
labels = ['Python','Java','c++','c']

plt.pie(sizes, labels=labels, colors=['#abcdef','#efd3ab','#aee6a3','#e6a1e6'])

plt.show()
```

# **Subplot**
## A subplot is a way to display multiple plots in a single figure, arranged in rows and columns.
## Syntax : plt.subplot(rows, columns, position)
## Use: It is used to compare multiple graphs side by side within the same figure.

```python
a = np.array([1,2,3,4,5])
b = np.array([3,8,1,10,2])
c = np.array([5,4,3,2,1])

# plot1
plt.subplot(1,2,1) # here (1,2,1) means 1row,2column,and this is first plot
plt.plot(a,b)

#plot2
plt.subplot(1,2,2)
plt.plot(a,c)

plt.show()
```

```python
a = np.array([1,2,3,4,5])
b = np.array([3,8,1,10,2])
c = np.array([5,4,3,2,1])

# plot1
plt.subplot(2,1,1) # here (2,1,1) means 2row,1column,and this is first plot
plt.plot(a,b)

#plot2
plt.subplot(2,1,2)
plt.plot(a,c)

plt.show()
```
